```python
from langchain_mistralai import ChatMistralAI
import os
```

```python
chatbot=ChatMistralAI(api_key=os.getenv('MISTRAL_API_KEY'),
                      temperature=0.3)
```

```python
def chatresponse(user_message:str)-> str:
    prompt=f"""
    you are chatbot assitant.you help the user with his below user message with your response

    user message:
    {user_message}
    """
    return chatbot.invoke(prompt).content

```

```python
from fastapi import APIRouter
from pydantic import BaseModel

class ChatRequest(BaseModel):
    message:str

router=APIRouter(prefix='/chat',tags=['chat'])

@router.post('/')
def chat(req:ChatRequest):
    reply=chatresponse(req.message)
    return {'reply':reply}
```

```python
from fastapi import FastAPI
app=FastAPI(title='AI Chatbot Assistant')
app.include_router(router)

@app.get("/")
def health():
    return {"status": "Chatbot running"}
```
